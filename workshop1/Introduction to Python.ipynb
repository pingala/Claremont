# Iterators

### * for loop: A control flow statement for specifying iteration, which allows code to be executed repeatedly.

### * while loop: A control flow statement for specifying iteration, which allows code to be executed repeatedly until a criterion is met.

### * nested loops: A loop of loops. That is, a loop inside of another loop.

```python
for animal in key: #colon is used to begin the loop
    print('I like ' + animal.lower() + "'s") #Turn everything to lower case
#Notice that we use tabs to denote the regions of the for loop that will be iteratively ran. 
#This works in all areas of python. You use tabs to denote what belongs to loops, functions, or classes.
```

```python
itr = 0
while itr < len(key):
    print('I like ' + key[itr] + "'s")
    itr += 1
    print('ITERATION COMPLETE: ' + str(itr))
print('All iterations ran.') #Not tabbed into the while loop statement. Therefore it is not repeated.
```

```python
for key in new_dict:
    print(new_dict[key])
```

```python
new_dict_sum = 0 #Create an object outside of the loop to store changes that occur in the loop
new_dict_str = '' #Create an empty string to build a string in the loop
for key in new_dict:
    if key is not 'Panda': #If the condition is True
        new_dict_str += key + ' + ' #Add the key to the string with a + sign following it
    else: #If the condition is False
        new_dict_str += key + ' = ' #If its the last key, add an equal sign instead of a + sign.
    new_dict_sum += new_dict[key] #Add all the numbers in the dictionary up.

new_dict_str + str(new_dict_sum)
```
